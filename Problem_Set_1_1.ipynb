# **Pb_1_Solutions_Python_Code**


```python
import numpy as np

# Function to determine the dimension of the span of the vectors
def determine_span(*vectors):
    matrix = np.column_stack(vectors)  # Combine vectors into a matrix
    rank = np.linalg.matrix_rank(matrix)  # Compute the rank of the matrix
    return rank

# (a) Vectors
v1_a = np.array([1, 2, 3])
v2_a = np.array([3, 6, 9])

# (b) Vectors
v1_b = np.array([1, 0, 0])
v2_b = np.array([0, 2, 3])

# (c) Vectors
v1_c = np.array([2, 0, 0])
v2_c = np.array([2, 2, 2])
v3_c = np.array([2, 2, 3])

# Determine the dimension of the span for each case
rank_a = determine_span(v1_a, v2_a)
rank_b = determine_span(v1_b, v2_b)
rank_c = determine_span(v1_c, v2_c, v3_c)

# Interpret the results
def interpret_span(rank):
    if rank == 1:
        return "Line"
    elif rank == 2:
        return "Plane"
    elif rank == 3:
        return "All of R^3"
    else:
        return "Invalid"

print("Geometric description for (a):", interpret_span(rank_a))
print("Geometric description for (b):", interpret_span(rank_b))
print("Geometric description for (c):", interpret_span(rank_c))
```

# **Pb_2_Solutions_Python_Code**


```python
import numpy as np

# Given diagonal vectors
v_plus_w = np.array([2, 3])
v_minus_w = np.array([6, -1])

# Solving for v and w using linear algebra
# Step 1: Solve for v by adding the two equations
v = (v_plus_w + v_minus_w) / 2

# Step 2: Solve for w by subtracting the two equations
w = (v_plus_w - v_minus_w) / 2

# Display the results
print(f"Vector v: {v}")
print(f"Vector w: {w}")
```

```python
# Fixing the issue and re-plotting without 'linestyles' for quiver
import numpy as np
import matplotlib.pyplot as plt

# Given diagonal vectors
v_plus_w = np.array([2, 3])
v_minus_w = np.array([6, -1])

# Solving for v and w using linear algebra
v = (v_plus_w + v_minus_w) / 2
w = (v_plus_w - v_minus_w) / 2

# Plotting the vectors and parallelogram
origin = np.array([0, 0])  # Origin point

# Create the parallelogram points
parallelogram = np.array([origin, v, v + w, w, origin])

# Set up the plot
plt.figure(figsize=(6, 6))
plt.grid(True)
plt.xlim(-4, 8)
plt.ylim(-4, 8)

# Plot vectors v, w, v-w, and v+w
plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1, color='b', label='v = (4, 1)')
plt.quiver(*origin, *w, angles='xy', scale_units='xy', scale=1, color='g', label='w = (-2, 2)')
plt.quiver(*origin, *v_plus_w, angles='xy', scale_units='xy', scale=1, color='m', label='v + w = (2, 3)')
plt.quiver(*origin, *v_minus_w, angles='xy', scale_units='xy', scale=1, color='c', label='v - w = (6, -1)')

# Plot the parallelogram
plt.plot(parallelogram[:, 0], parallelogram[:, 1], 'r--', label='Parallelogram')

# Plot labels and legend
plt.title('Parallelogram with vectors v, w, v + w, and v - w')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given diagonal vectors
v_plus_w = np.array([5, 1])
v_minus_w = np.array([1, 5])

# Solving for v and w using linear algebra
v = (v_plus_w + v_minus_w) / 2
w = (v_plus_w - v_minus_w) / 2

# Plotting the vectors
origin = np.array([0, 0])  # Origin point

# Set up the plot
plt.figure(figsize=(6, 6))
plt.grid(True)
plt.xlim(-4, 6)
plt.ylim(-4, 6)

# Plot vectors v and w
plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1, color='b', label='v = (3, 3)')
plt.quiver(*origin, *w, angles='xy', scale_units='xy', scale=1, color='g', label='w = (2, -2)')

# Plot labels and legend
plt.title('Vectors v and w')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np

# Coefficients matrix (from the system of equations)
A = np.array([[2, -1, 0],
              [-1, 2, -1],
              [0, -1, 2]])

# Right-hand side vector b
b = np.array([1, 0, 0])

# Solving the system of equations
solution = np.linalg.solve(A, b)

# Display the values of c, d, e
c, d, e = solution
print(f"c = {c}, d = {d}, e = {e}")
```
